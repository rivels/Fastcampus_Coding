```python
class Human():
    health = 40
    attack_power = 10
    defence = 5
    def steam_pack(self):
        self.attack_power *= 2
class Orc():
    health = 60
    attack_power = 5
    defence = 10
    def shield(self):
        self.defence *= 2
        
class wizzard(Human):
    def spec(self):
        self.attack_power += 10
        
class warrior(Orc):
    def spec(self):
        self.health += 10
```

```python
def disp_var(obj):
    return [var for var in dir(obj) if var[:2] != "__"]
```

```python
owa.health, owa.attack_power, owa.defence,disp_var(owa)
```

# super
- 부모 클래스의 메서드 코드를 가져올때 사용

```python
class Human:
    def __init__(self):
        self.health = 40
    def move(self):
        print('move')
```

```python
class Marine(Human):
    def __init__(self):
        self.attack_power = 5
        super(Marine,self).__init__()
    def attack(self):
        print('attack')
```

```python
class Calculator:
    
    def __init__(self,num1=1,num2=2):
        self.num1 = num1
        self.num2 = num2
        
    def setter_num1(self,num1):
        print('setter')
        if isinstance(num1,(int,float)):
            self.num1 = num1
        else:
            print('wrong data type(only int, float)')
        
    def getter_num1(self):
        print('getter')
        datas = {1:'one',2:'two',3:'three'}
    def plus(self):
        return self.num1 + self.num2
    
    num1 = property(getter_num1,setter_num1)
```

```python
calc = Calculator()
```

# 클래스 메서드 종류
- instance method
    - 인스턴스(객체)를 통해서 호출
    - 파라미터에 self를 사용
- class method
    - 클래스를 통해서 호출
    - 파라미터에 cls를 사용
- static method
    - 객체를 생성하지 않아도 바로 클래스를 통해서 일반 함수처럼 메서드 사용가능
    - 일반함수와 같지만 위치만 클래스 내부
    

```python
# 계좌(account) class
```

```python
class Account:
    interest = 1.01
    
    def __init__ (self, asset = 10000):
        self.asset = asset
    def deposit(self, amount): #입금
        self.asset += amount
    def withdraw(self, amount): #출금
        if self.asset < amount:
            print('잔액이 {}원 부족합니다'.format(amount-self.asset))
        else:
            self.asset -= amount
            print('{}원이 출금되었습니다.'.format(amount))
    def add_interest(self): #자산에 이자 추가
        self.asset = int(self.asset * self.interest)
        
    def ins_change_interest(self, interest):    
        self.interest = interest

    @classmethod
    def cls_change_interest(cls, interest):
        cls.interest = interest
    
    def interest_grade(interest):
        if interest >= 1.05:
            print('high interest')
        else:
            print('low interest')
        
```

```python
a1 = Account(10000)
a2 = Account(20000)
a3 = Account(30000)
```

```python
a1.interest, a2.interest, a3.interest
```

```python
a1.add_interest()
```

```python
a1.asset
```

```python
# 클래스 메서드의 사용
Account.cls_change_interest(1.05)
```

```python
# 스태틱 메서드의 사용
Account.interest_grade(1.06)
```

```python
#2.특별한기능
# str : 객체를 print 했을때 출력되는 데이터
# repr : ipython에서 출력하는 데이터
```

```python
class Number:
    def __init__(self,number):
        self.number = number
    def __add__(self,obj):
        return self.number - obj.number
    def __repr__(self): #개발자용으로 출력해서 객체를 확인할때
        return "<Number number:{}>".format(self.number)
    def __str__(self):
        return str(self.number)
```

```python
n1 = Number(3)
```
