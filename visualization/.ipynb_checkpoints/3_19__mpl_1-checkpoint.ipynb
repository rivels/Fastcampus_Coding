```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
datas = np.random.randint(10,size=10)
datas
```

```python
plt.plot(datas)
plt.show()
```

```python
# x값 설정
x_datas = np.arange(0,200,20)
y_datas = datas
```

```python
len(x_datas), x_datas, len(y_datas),y_datas
```

```python
plt.plot(x_datas,y_datas)
plt.show()
```

```python
plt.scatter(x_datas,y_datas)
plt.show()
```

## 스타일설정
- 표현식
- 직접 스타일을 아규먼트로 설정

- 표현식 설정 : 색상, 마커스타일, 라인스타일 

```python
plt.plot(y_datas,"bv--") # blue//삼각형//점선

#color : b lue, g reen, r ed, y ellow
# . : point, o : circle, ^ : traiangle, * : star
# -- : 실선의 점선, : = 점선, 
```

### 직접 스타일을 아규먼트로 설정
- c : 선색상, lw : 선굵기, ls : 선스타일, marker : 마커종류
- ms: 마커 크기, mec : 마커선색상, mew : 마커선 굵기, mfc: 마커내부색상

```python
plt.plot(y_datas, c="orange",lw=5,ls="--",marker='o', ms = 10, mec="black" , mew = 2, mfc ="yellow")
```

```python
# x축, y축 범위(limit) 설정
```

```python
plt.plot(y_datas,"bv--")
plt.xlim(-2,12) # 그래프를 어디부터 표시할지 설정
plt.ylim(-2,20)
plt.show()
```

# tick 설정

```python
plt.plot(y_datas,"bv--")
plt.xticks(range(0,10,3)) # range로 그래프의 표시 단위 설정가능
plt.yticks(range(0,10,1))
plt.show()
```

```python
# 멀티라인그리기
```

```python

datas = [np.random.randint(5, size=5) for _ in range(3)]
datas
```

```python
# plot함수 한번 호출
x = range(5)
plt.plot(x,datas[0], "b^:", x,datas[1],"ro--", x,datas[2],"yv-")
```

```python
# plot함수 세번 호출
```

```python
# plot 함수 세번 호출
for idx in range(3):
    plt.plot(x, datas[idx])
plt.show()
```

```python
x, datas[0]
```

```python
plt.plot(x,datas[0])
```

```python
datas = [np.random.randint(5, size=5) for _ in range(3)]
for idx in range(3):
    plt.plot(x, datas[idx], label=f"datas{idx}")
plt.legend(loc=1) # 범례표시하는 명령어// loc는 범례의 위치 설정//아래그래프는 1번위치(right upper)
plt.show()
```

```python
# x라벨 y라벨은 그래프 옆에 띄워지도록 하는 변수
plt.plot(x,datas[0], "b^:", x,datas[1],"ro--", x,datas[2],"yv-")
plt.title("graph")
plt.xlabel("xxxxx")
plt.ylabel("yyyyy")
```

```python
### 그래프 크기 조절 : figure(figsize =(x,y))
```

```python
plt.figure(figsize=(10,10)) 
plt.plot(y_datas)
plt.show()
```

그래프여러개 그리기

```python
plt.figure(figsize=(10,10)) 
plt.subplot(121) # 2*1 행렬에서 1번째
plt.plot(datas[0])
plt.subplot(122) # 2*1 행렬에서 2번째
plt.plot(datas[1])
```

```python
# for문 이용해서 차트 반복문으로 그려내기
datas = [np.random.randint(5, size=5) for _ in range(4)]
print(datas)
plt.figure(figsize=(20,10))

for idx in range(4):
    plt.subplot(221+idx)
    plt.plot(datas[idx])
plt.show()
```
