```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# bar
datas = np.random.randint(10,size=5)
plt.bar(range(len(datas)),datas, width =0.3)
plt.show()
```

```python
# rgba() : 16진수를 씀 0~F //  red, green, blue, 투명도
datas = np.random.randint(5, 10,size=5)
error = np.random.randint(1,5,size=5)
plt.bar(range(len(datas)),datas, width =0.3, alpha=0.27, yerr=error)
plt.show()
```

```python
# rgba() : 16진수를 씀 0~F //  red, green, blue, 투명도
datas = np.random.randint(5, 10,size=5)
error = np.random.randint(1,5,size=5)
plt.barh(range(len(datas)),datas, alpha=0.27, xerr=error) # xerr 는 오차를 설정할때
plt.show()
```

# seaborn : 가장 자주 사용됨
- matplotlib 기반의 python 시각화 라이브러리
- 고급 인터페이스를 제공

```python
import seaborn as sns
```

```python
# 아래와 같이 입력해야 style이 적용됨
sns.set()
```

```python
sns.set_style("whitegrid")
plt.plot(datas)
plt.show()
```

```python
sns.set_theme(style="ticks")

rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x=x, y=y, kind="hex", color="#4CB391")
```

```python
sns.palplot(sns.color_palette()) #팔레트 : 첫번째 그래프는 파랑, 두번째는 주황 순으로 색이 정해짐을 나타냄
```

```python
# 컬러 테마 변경
theme = sns.color_palette("vlag",20)
sns.palplot(theme)
theme
```

```python
sns.set_palette("husl",10)
```

```python
sns.set_style("whitegrid")
for i in range(10):
    plt.plot(np.random.randint(5,size=5))
plt.show()
```

```python
# examples
```

```python
# iris
iris = sns.load_dataset("iris")
iris.tail(2)
```

```python
# lmplot
```

```python
sns.lmplot("sepal_length","sepal_width",data=iris, hue="species", fit_reg=True) #fit_reg 는 선형회귀의 선을 넣을지 여부
# hue는 타겟 컬럼 데이터
plt.show()
```

```python
sns.pairplot(iris, hue="species") # 각 feature들의 경우의수에따라 그래프를 모두 그려줌
plt.show()
```

```python
# flights : 여객기 승객수
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
df = flights.pivot('month','year',"passengers")
```

```python
#heatmap : 위의 피봇처럼 만들때 숫자가 크면 색이 진해지도록 표시
```

```python
sns.heatmap(df,annot=True, fmt='d', cmap='YlGn')
plt.show()
```

```python

sns.heatmap(iris.corr(),annot=True,  cmap='YlGn')
plt.show()
```
